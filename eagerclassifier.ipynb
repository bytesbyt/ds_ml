```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
# load the dataset
data_food = pd.read_csv('food_data.csv')
data_food.head()
```

```python
# selecting the independatn and dependant variable
data_food.index = data_food.iloc[:, 0]
X = data_food.iloc[:, 1:16]
y = data_food.iloc[:, 16]
```

```python
X
```

```python
y
```

```python
# split our dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state= 32)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
# make prediction
y_pred_gnb = clf.predict(X_test)
y_pred_gnb
```

```python
# put in a dataframe
pd.DataFrame({'Actual_food':y_test, 'Predicted_food': y_pred_gnb})
```

```python
# model evaluation
# accuracy score
metrics.accuracy_score(y_test, y_pred_gnb)
```

```python
# logistic regression
# load the dataset
data_candy = pd.read_csv('candy-data.csv')
data_candy.head()
```

```python
X = data_candy.iloc[:, [3, 12]]
y = data_candy.iloc[:, 1]
```

```python
#data spliting
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=32)
```

```python
x_train
```

```python
# standarize the data point
# standard scaler
# mean of 0 and a standard deviation of 1
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
# logistic regression
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
# make prediction
y_pred_log = log_reg.predict(x_test)
```

```python
y_pred_log
```

```python
pd.DataFrame({'Actual_cnady_type': y_test, 'Predicted_candy_type': y_pred_log})
```

```python
#model evaluation
metrics.accuracy_score(y_test, y_pred_log)
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred_log)
```
